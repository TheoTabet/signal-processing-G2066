```python
import itertools
import numpy as np
import pandas as pd

def compute_final_number(sets, fixed_number=None, target=500):
 
    # Generate all combinations: one number from each set
    all_combinations = list(itertools.product(*sets))
    
    results = []
    for combo in all_combinations:
        product = np.prod(combo)
        if fixed_number is not None:
            product *= fixed_number
        # Compute final 6th number
        final_number = target / product
        results.append(list(combo) + ([fixed_number] if fixed_number is not None else []) + [final_number])
    
    # Build DataFrame
    col_names = [f'Set{i+1}' for i in range(len(sets))]
    if fixed_number is not None:
        col_names.append('Fixed')
    col_names.append('FinalNumber')
    df = pd.DataFrame(results, columns=col_names)
    return df


# Example usage

if __name__ == "__main__":
    # 5 sets of numbers (example)
    set1 = [3.7, 3.4, 2]
    set2 = set1
    set3 = set1
    set4 = set1
   

    sets = [set1, set2, set3, set4]  # only the 4 variable sets
    fixed_number = 12  # this is the number that's constant
    target = 500

    df = compute_final_number(sets, fixed_number=fixed_number, target=target)

    # Save to Excel
    df.to_excel("permutation_costs.xlsx", index=False)
    print("Results saved to permutation_costs.xlsx")
    print(df.sum())
```
